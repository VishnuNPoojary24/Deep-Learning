```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Placement.csv")
```

```python
df=pd.DataFrame(df)
```

```python
df.head()
```

```python
df.drop(columns=['Student_ID'],axis=1,inplace=True)
```

```python
df.head()
```

```python
from sklearn.linear_model import Perceptron
```

```python
per=Perceptron()
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
per.fit(X,y)
```

```python
per.coef_
```

```python
per.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values,y.values,clf=per,legend=2)
```

```python
sns.scatterplot(x='CGPA',y='IQ',data=df,hue=df['Placement'])
```

```python
#after standardizing the input columns 
from sklearn.preprocessing import StandardScaler

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Refit the model with standardized features
per.fit(X_scaled, y)

# Plot decision regions
plot_decision_regions(X_scaled, y.values, clf=per, legend=2)
```
